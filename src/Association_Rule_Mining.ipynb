# load data preproessed_RFM

```python
data = pd.read_csv(r"D:\Sem 4\Rekomendasi Sistem\Tugas\PA\data\Preprocessed_RFM.csv")
data.head()
```

# Basket Matrix

```python
basket = pd.crosstab(
    data['CustomerID'],
    data['ProductCategory']
)
```

```python
basket.head()
```

# One Hot Encoding

```python
basket = (basket > 0).astype(int)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(
    basket,
    min_support=0.05,
    use_colnames=True
)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(
    basket,
    min_support=0.3,
    use_colnames=True
)

frequent_itemsets.sort_values(
    'support',
    ascending=False
).head()
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(
    frequent_itemsets,
    metric='support',
    min_threshold=0.01
)
rules.head()
```

```python
rules = rules[
    (rules['confidence'] >= 0.8) &
    (rules['lift'] >= 1)
]
```

```python
rules.to_csv(
    'D:\Sem 4\Rekomendasi Sistem\Tugas\PA\data\processed_association_rules.csv',
    index=False
)
```
