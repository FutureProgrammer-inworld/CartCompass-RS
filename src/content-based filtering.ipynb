# library

```python
import pandas as pd
```

# load data

```python
data =  pd.read_csv(r"D:\Sem 4\Rekomendasi Sistem\Tugas\PA\data\processed_customer_segmented.csv")
data.head()
```

# Buat Profile

```python
data['Profile'] = (
    data['Gender'].astype(str) + ' ' +
    data['AgeGroup'].astype(str) + ' ' +
    data['Region'].astype(str) + ' ' +
    data['FavoriteCategory'].astype(str) + ' ' +
    data['Segment'].astype(str)
)
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(
    data['Profile']
)
```

# Similarity Matrix

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(
    tfidf_matrix
)
```

```python
import joblib

joblib.dump(
    similarity_matrix,
    'customer_similarity.pkl'
)
```
