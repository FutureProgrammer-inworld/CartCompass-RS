# library

```python
import pandas as pd
```

```python
data = pd.read_csv(r"D:\Sem 4\Rekomendasi Sistem\Tugas\PA\data\processed_customer_profile.csv")
data.head()
```

```python
# recency
data['R_Score'] = pd.qcut(
    data['Recency'],
    q=4,
    labels=[4,3,2,1]
)
```

```python
# Frequency
data['F_Score'] = pd.qcut(
    data['Frequency'],
    q=4,
    labels=[1,2,3,4]
)
```

```python
# monetary
data['M_Score'] = pd.qcut(
    data['Monetary'],
    q=4,
    labels=[1,2,3,4]
)
```

```python
data.head()
```

```python
# Membuat kolom RFM_Score dengan menggabungkan R_Score, F_Score, dan M_Score 
data['RFM_Score'] = (
    data['R_Score'].astype(str)
    + data['F_Score'].astype(str)
    + data['M_Score'].astype(str)
)
```

```python
# membuat rules untuk segmentasi pelanggan berdasarkan RFM_Score    
def segment_customer(row):

    r = int(row['R_Score'])
    f = int(row['F_Score'])
    m = int(row['M_Score'])

    if r >= 3 and f >= 3 and m >= 3:
        return "Loyal Customer"

    elif r >= 2 and f >= 2:
        return "Potential Customer"

    elif r == 1:
        return "At Risk Customer"

    else:
        return "New Customer"
```

```python
data['Segment'] = data.apply(
    segment_customer,
    axis=1
)
```
